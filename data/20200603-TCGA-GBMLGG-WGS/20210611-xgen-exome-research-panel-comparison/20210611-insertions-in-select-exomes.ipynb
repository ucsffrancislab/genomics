```python
import sqlite3
sqlite3.version
```

```python
exomes=pd.read_csv("xgen-exome-research-panel-v2-targets-hg38.bed",header=None,sep="\t",
 names=['chromosome','start','stop','exome','unknown','direction'])
```

```python
exomes.head()
```

```python
points=pd.read_csv("HERVK113.paired.Q20.all_insertion_points.uniqc.txt",header=None,sep="\s+",
 names=['count','chromosome','position'])
```

```python
points.head()
```

```python
conn = sqlite3.connect(':memory:')
exomes.to_sql('exomes', conn, index=False)
points.to_sql('points', conn, index=False)
qry = '''
    select *
    from
        exomes join points on
        exomes.chromosome = points.chromosome and
        points.position between exomes.start and exomes.stop
    '''
df = pd.read_sql_query(qry, conn)
df
```
