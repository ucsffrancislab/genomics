# Read counts

```python
# header uses subject and not sample
#rc=pd.read_csv("data/TCGA_raw_counts.csv",header=0, names=['sample','count'])
rc=pd.read_csv("data/TCGA_raw_counts.csv")
rc.rename(columns={"subject": "sample"},inplace=True)
#rc['subject']=rc['sample'].astype(str).str[0:7]
rc.head()
```
