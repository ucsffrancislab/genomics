```python
df = pd.read_csv('feature_counts.5.cutadapt2.mirna.summary.tsv',sep="\t")
sum=df.sum().to_frame().T
sum['Status']='sum'
sum
df=pd.concat([df,sum],axis='index')
del(sum)
df.set_index('Status',inplace=True)
df=df.T
df.drop(['Unassigned_Read_Type','Unassigned_Singleton','Unassigned_MappingQuality','Unassigned_Chimera',
    'Unassigned_FragmentLength','Unassigned_Duplicate','Unassigned_MultiMapping','Unassigned_Secondary',
    'Unassigned_NonSplit','Unassigned_Overlapping_Length',
    'Unassigned_Unmapped','Unassigned_NoFeatures','Unassigned_Ambiguity'],axis='columns',inplace=True)
#df['assigned per million']=1e6*df['Assigned']/df['sum']
cats1=df
del(df)
#cats1



df = pd.read_csv('feature_counts.5.cutadapt2.mirna.tsv',sep="\t",header=1)

df.set_index('Geneid',inplace=True)
df.drop(['Chr','Start','End','Strand','Length'],axis='columns',inplace=True)
df[df>0]=1

cats2=df.sum().to_frame()
del(df)
cats2.rename(columns={0: "unique"},inplace=True)
#cats2['library']='CATS'
cats2

cats=pd.concat([cats1,cats2],axis='columns')

cats['library']='CATS'

#cats=cats_LTR+cats_Retroposon
cats['assigned per million']=1e6*cats['Assigned']/cats1['sum']
cats['unique per million']=1e6*cats['unique']/cats['sum']

#cats.drop(['SFHH005ag','SFHH005ar','SFHH005k','SFHH005v'],inplace=True,axis='index')
cats.drop(['SFHH005ag','SFHH005k'],inplace=True,axis='index')
cats
```

```python
df = pd.read_csv('feature_counts.6.cutadapt2.mirna.summary.tsv',sep="\t")
sum=df.sum().to_frame().T
sum['Status']='sum'
sum
df=pd.concat([df,sum],axis='index')
del(sum)
df.set_index('Status',inplace=True)
df=df.T
df.drop(['Unassigned_Read_Type','Unassigned_Singleton','Unassigned_MappingQuality','Unassigned_Chimera',
    'Unassigned_FragmentLength','Unassigned_Duplicate','Unassigned_MultiMapping','Unassigned_Secondary',
    'Unassigned_NonSplit','Unassigned_Overlapping_Length',
    'Unassigned_Unmapped','Unassigned_NoFeatures','Unassigned_Ambiguity'],axis='columns',inplace=True)
#df['assigned per million']=1e6*df['Assigned']/df['sum']
lex1=df
del(df)
#lex1

df = pd.read_csv('feature_counts.6.cutadapt2.mirna.tsv',sep="\t",header=1)

df.set_index('Geneid',inplace=True)
df.drop(['Chr','Start','End','Strand','Length'],axis='columns',inplace=True)
df[df>0]=1

lex2=df.sum().to_frame()
del(df)
lex2.rename(columns={0: "unique"},inplace=True)
#lex2['library']='CATS'
lex2

lex=pd.concat([lex1,lex2],axis='columns')


#lex=lex_LTR+lex_Retroposon
lex['library']='Lexogen'
lex['assigned per million']=1e6*lex['Assigned']/lex['sum']
lex['unique per million']=1e6*lex['unique']/lex['sum']
lex.drop(['SFHH006ag','SFHH006k'],inplace=True,axis='index')
lex
```

```python
df=pd.concat([cats,lex])
df
```

```python
df['assigned per million']=df['assigned per million'].astype(float)
df['unique per million']=df['unique per million'].astype(float)
```

```python
ax=sns.boxplot(x="library", y="assigned per million", data=df)
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Assigned per million")
plt.show()
```

```python
ax=sns.boxplot(x="library", y="unique per million", data=df)
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique per million")
plt.show()
```
