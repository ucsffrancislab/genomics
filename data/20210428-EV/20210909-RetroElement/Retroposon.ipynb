```python
df = pd.read_csv('feature_counts.5.cutadapt2.Retroposon.summary.tsv',sep="\t")
df
```

```python
sum=df.sum().to_frame().T
sum['Status']='sum'
sum
```

```python
df=pd.concat([df,sum],axis='index')
del(sum)
df
```

```python
df.set_index('Status',inplace=True)
df=df.T
df.drop(['Unassigned_Read_Type','Unassigned_Singleton','Unassigned_MappingQuality','Unassigned_Chimera',
    'Unassigned_FragmentLength','Unassigned_Duplicate','Unassigned_MultiMapping','Unassigned_Secondary',
    'Unassigned_NonSplit','Unassigned_Overlapping_Length',
    'Unassigned_Unmapped','Unassigned_NoFeatures','Unassigned_Ambiguity'],axis='columns',inplace=True)
df
```

```python
df['assigned per million']=1e6*df['Assigned']/df['sum']
df
```
