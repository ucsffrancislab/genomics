```python
df = pd.read_csv('feature_counts.5.cutadapt2.mirna.tsv',sep="\t",header=1)

df.set_index('Geneid',inplace=True)
df.drop(['Chr','Start','End','Strand','Length'],axis='columns',inplace=True)
df[df>0]=1

cats=df.sum().to_frame()
del(df)
cats.rename(columns={0: "unique"},inplace=True)
cats['library']='CATS'
cats
```

```python
df = pd.read_csv('feature_counts.6.cutadapt2.mirna.tsv',sep="\t",header=1)

df.set_index('Geneid',inplace=True)
df.drop(['Chr','Start','End','Strand','Length'],axis='columns',inplace=True)
df[df>0]=1

lex=df.sum().to_frame()
del(df)
lex.rename(columns={0: "unique"},inplace=True)
lex['library']='Lexogen'
lex
```

```python
ax=df.boxplot(column='unique',by=['library'])
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique Hits")
plt.title('')
plt.suptitle('')
plt.show()
```

```python
ax=sns.boxplot(x="library", y="unique", data=df)
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique Hits")
plt.show()
```

```python
ax=sns.swarmplot(x="library", y="unique", data=df,size=5,
    edgecolor='black',linewidth=1,linestyle='-')
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique Hits")
plt.show()
```

```python
ax=sns.boxplot(x="library", y="unique", data=df)
ax=sns.swarmplot(x="library", y="unique", data=df,size=5,
    edgecolor='black',linewidth=1,linestyle='-')
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique Hits")
plt.show()
```

```python
ax=sns.stripplot(x="library", y="unique", data=df,size=5,
    edgecolor='black',linewidth=1,linestyle='-')
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique Hits")
plt.show()
```

```python
ax=sns.violinplot(x="library", y="unique", data=df)
ax.set_xlabel("Preparation Library")
ax.set_ylabel("Unique Hits")
plt.show()
```
